```python
nulos = df.isnull().sum()
print('======== Porcentaje valores nulos ========')
resultado = (nulos[nulos > 0] / df.shape[0] *
             100).sort_values(ascending=False).round(2)
print(resultado.astype(str) + '%')
```

```python
EstacionamientosPorcentaje = (df['Parking'].value_counts(
    dropna=False).sort_index() / df.shape[0] * 100).round(2)
print(EstacionamientosPorcentaje.astype(str) + '%')
```
